```python
import plotly.plotly as py
import plotly.graph_objs as go

import pandas as pd

df = pd.read_csv("price_and_position_data\\price_bt_yeni.csv", delimiter=',', header=0)


trace_high = go.Scatter(
                x=df['Date'],
                y=df['Price'],
                name = "Price",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

trace2 = go.Bar(
    x=['2011-4-1', '2011-5-1', '2012-1-1'],
    y=[200, 300, 400],
    
)

data = [trace_high, trace2]

layout = dict(
    title = "Manually Set Date Range",
    xaxis = dict(
                 tick0='2011-4-1',
                dtick='M3'
               ),
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Manually Set Range")
```

```python
import csv

def format_date_string_plotly(unformatted_date):
    formatted_date = ""
    year = unformatted_date[unformatted_date.rfind(".")+1:]
    month = unformatted_date[unformatted_date.rfind(".", 0, unformatted_date.rfind("."))+1:unformatted_date.rfind(".")]
    month = month.zfill(2)
    day = unformatted_date[:unformatted_date.find(".")]
    day = day.zfill(2)
    formatted_date_key = year + "-" + month + "-" + day
    return formatted_date_key

listoflists = []


with open('demo\\datequartmap.csv', 'rb') as csvfile:
    prices = csv.reader(csvfile, delimiter=';', quotechar='"')
    for row in prices:
        listtoadd = [row[0], row[1], format_date_string_plotly(row[2])]
        listoflists.append(listtoadd)



with open('demo\\datequartmap_yeni.csv', 'wb') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=';')
    for v in listoflists:
        spamwriter.writerow([v[0], v[1], v[2]])
print 'Done!'
```

```python
import csv

def format_date_string_plotly(unformatted_date):
    formatted_date = ""
    year = unformatted_date[unformatted_date.rfind(".")+1:]
    month = unformatted_date[unformatted_date.rfind(".", 0, unformatted_date.rfind("."))+1:unformatted_date.rfind(".")]
    month = month.zfill(2)
    day = unformatted_date[:unformatted_date.find(".")]
    day = day.zfill(2)
    formatted_date_key = year + "-" + month + "-" + day
    return formatted_date_key

dictr = {}

with open('demo\\price_ses.csv', 'rb') as csvfile:
    prices = csv.reader(csvfile, delimiter=';', quotechar='"')
    for row in prices:
        dictr[format_date_string_plotly(row[0])] = row[1]



with open('demo\\price_ses_yeni.csv', 'wb') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=';')
    for v in sorted(dictr.iterkeys()):
        spamwriter.writerow([v, dictr[v]])
print 'Done!'
```
